```python
%matplotlib inline

vendas = [100, 400, 700, 800]

produtos = ['Blusa', 'Bone', 'Chinelo', 'Bermuda' ]

plt.pie(vendas, labels=produtos)
plt.show()
```

```python
%matplotlib inline 

vendas = [100, 200, 300, 500]
produto = ['Apple', 'Samsung', 'Asus', 'Dell']
#definindo a serabilidade dos espaços do gráficos
explode = (0.1, 0,0,0)

plt.pie(vendas, labels=produto, autopct='%1.1f%%', shadow=True, explode=explode)
plt.legend(produto, loc=2)
plt.axis('equal')
plt.show()
```

 # subplots métedo avançado 

```python
x = [1,2,4,5,6]
y = [1,2,3,4,6]
# criar um objeto do tipo figure 
fig = plt.figure()

#define o tamanho em percentual de um gráfico 
# altura e largura do eixo da figura 

axes_a = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_b = fig.add_axes([0.2,0.5,0.3,0.3])

#plotar a figura com dois eixos. 
# o 'r--' siginifica linha tracejada em vermelho(r)

axes_a.plot(x,y)
axes_b.plot(x,y, 'r--')
```

```python
 x = [1,2,3,4]
 y = [10,20,25,40]


#alterar a largura e altura do gráfico (width, height)

fig, ax = plt.subplots(figsize=(20,5))

ax.plot(x,y, 'r--')
```

```python
# define as configuração dos plots 
# cada plot tera o mesmo tamanhos das figuras 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))


# dados para cada subplot 

ax1.bar([1,2,3], [3,4,5])
ax2.barh([0.5, 1,2.5], [0,1,2])
plt.show()
```

# define configuração dos eixos com meth 'set()"

```python
fig, (ax1, x2) = plt.subplots(1,2 ,figsize=(10,5))

ax1.bar([1,2,3], [3,4,5], color='r')
ax2.barh([0.5,1,2.5], [0,1,2,])

ax1.set(title="grafica de barras veticais" ,xlabel='Eixo x ', ylabel='Eixo y')
ax2.set(title='Barras horizontais', xlabel='Eixo x ', ylabel='Eixo y')
plt.show()
```
